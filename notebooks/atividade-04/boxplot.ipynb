# Análise exploratória de dados

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
ax=sns.boxplot( x='day', y='total_bill', data=df )
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(6,3))
ax=sns.boxplot(data=df,x='day', y='total_bill', color='cornflowerblue')
plt.ylabel('Conta(USD)')
plt.xlabel('Dia da Semana')
plt.title('Distribuição de gorjetas ao longo da semana')
plt.show()
```

```python
ax.bar_label(ax.containers[0], fmt='%.1f', label_type='center', color='white')
```

```python
plt.xlabel('Arrecadação (milhões de dólares)')
plt.ylabel('Filme')
plt.title('Arrecadação de Filmes')
```

```python
ax.set(frame_on=False)
ax.axvline(0, color="k", clip_on=False)
plt.show()
```
