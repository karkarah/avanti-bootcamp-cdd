## Descrição dos dados usando estatísitca

### Variaveis Qualitativas

```python
# Qualitativas 
df_qual=df_dict.query('Tipo == "Qualitativa"')
qualitativas = list(df_qual['Variável'])

# Quantitativas
df_qual=df_dict.query('Tipo == "Quantitativa"')
quantitativas = list(df_qual['Variável'])

#Moda
#print(df['sex'].mode())
print('moda: ')
df[qualitativas].mode()
```

```python
# Loop externo 
for coluna in qualitativas:    
    # cada valor únicos na coluna
    unicos = df[coluna].unique()
    # Loop interno nos valores
    for valor in unicos:
        # Filtrar no valor atual 
        filtrado = df[df[coluna] == valor]
        # Descrever os dados 
        print(coluna,' - ',valor)
        print(df[df[coluna] == valor].describe())
        print()
        
```
